### Filter pada Numpy Array 1D

```python
import numpy as np

arr = np.array([41, 42, 43, 44])
x = [True, False, True, False]
arr = arr[x]

print(arr)
```

```python
arr = np.array([41, 42, 43, 44])
filter_arr = arr > 42
arr = arr[filter_arr]

print(arr)
```

```python
arr = np.array([1, 2, 3, 4, 5, 6, 7])
filter_arr = arr % 2 == 0
arr = arr[filter_arr]

print(arr)
```

---

### Filter pada Numpy Array 1D menggunakan lebih dari 1 kondisi

```python
arr = np.array([1, 2, 3, 4, 5, 6, 7])
filter_arr = (arr % 2 == 0) & (arr > 2)
arr = arr[filter_arr]

print(arr)
```

### Filter pada Numpy Array 2D

```python
arr = np.array([['1', 'one'],
                ['2', 'two'],
                ['3', 'three'],
                ['4', 'four'],
                ['5', 'five']])
 
filter = np.array(['two', 'four'])
arr = arr[np.in1d(arr[:, 1], filter)]

print(arr)
```
