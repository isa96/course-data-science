```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
plt.style.use('seaborn') # plt.style.available
```

### Membuat diagram sederhana dengan Seaborn

```python
data = sns.load_dataset("iris") 
  
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
```

### Membuat diagram sederhana dengan Seaborn & Pyplot

```python
data = sns.load_dataset("iris") 

plt.figure(figsize=(10, 12))

sns.set_palette('vlag')
plt.subplot(211)
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title('Title using Matplotlib Function')
plt.plot()

sns.set_palette('Accent')
plt.subplot(212)
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title('Title using Matplotlib Function')
plt.plot()

plt.show()
```

### Membuat Scatter Plot

```python
data = sns.load_dataset("iris")
  
sns.scatterplot(x='sepal_length', y='sepal_width', data=data)
plt.show()
```

### Membuat Bar Chart

```python
data = sns.load_dataset("iris")
  
sns.barplot(x='species', y='sepal_length', data=data)
plt.show()
```

```python
data = sns.load_dataset("iris")
  
sns.countplot(x='species', data=data)
plt.show()
```

### Membuat Histogram

```python
data = sns.load_dataset("iris")

sns.histplot(data['sepal_width'])
plt.show()
```

### Membuat Box Plot

```python
data = sns.load_dataset("iris")
  
sns.boxplot(x='species', y='sepal_width', data=data)
plt.show()
```

### Membuat Violin Plot

```python
data = sns.load_dataset("iris")
  
sns.violinplot(x='species', y='sepal_width', data=data)
plt.show()
```

### Membuat Pair Plot

```python
data = sns.load_dataset("iris")
  
sns.pairplot(data=data, hue='species')
plt.show()
```

### Membuat Heatmap

```python
data = sns.load_dataset("iris")
  
tc = data.corr()
  
sns.heatmap(tc)
plt.show()
```
