### Membuat dua plot secara horizontal

```python
# Plot 1
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 10, 80])

plt.subplot(1, 2, 1) # 1 row 2 columns
plt.plot(x,y)

# Plot 2
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 90, 120])

plt.subplot(1, 2, 2) # 1 row 2 columns
plt.plot(x,y)

plt.show()
```

### Membuat dua plot secara vertikal

```python
# Plot 1
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 10, 80])

plt.subplot(2, 1, 1)  # 2 rows 1 column
plt.plot(x,y)

# Plot 2
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 90, 120])

plt.subplot(2, 1, 2) # 2 rows 1 column
plt.plot(x,y)

plt.show()
```

### Memberi title pada dua plot

```python
# Plot 1
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 10, 80])

plt.subplot(1, 2, 1) # 1 row 2 columns
plt.plot(x,y)
plt.title('SALES')

# Plot 2
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 90, 120])

plt.subplot(1, 2, 2) # 1 row 2 columns
plt.plot(x,y)
plt.title('INCOME')

plt.show()
```

### Memberi title utama pada dua plot

```python
plt.figure(figsize=(14, 4))

# Plot 1
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 10, 80])

plt.subplot(1, 2, 1) # 1 row 2 columns
plt.plot(x,y)
plt.title('USERS')

# Plot 2
x = np.array([0, 1, 2, 3])
y = np.array([30, 80, 90, 120])

plt.subplot(1, 2, 2) # 1 row 2 columns
plt.plot(x,y)
plt.title('INCOME')

plt.suptitle('TRACTION JUNE 2022')
plt.show()
```
