### Membuat diagram sederhana dengan Pyplot

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
xpoints = np.array([0, 20])
ypoints = np.array([0, 180])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([0, 20])
ypoints = np.array([180, 0])

plt.plot(xpoints, ypoints)
plt.show()
```

### Membuat diagram sederhana dengan marker value

```python
xpoints = np.array([0, 20])
ypoints = np.array([0, 180])

plt.plot(xpoints, ypoints, marker='o')
plt.show()
```

```python
plt.plot(xpoints, ypoints, marker='*')
plt.show()
```

### Membuat diagram banyak titik dengan Pyplot

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```
