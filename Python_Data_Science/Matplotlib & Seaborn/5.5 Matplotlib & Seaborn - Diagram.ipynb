```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

### Membuat Scatter Plot

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])

plt.scatter(x, y)
plt.show()
```

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])

plt.scatter(x, y)

x = np.array([2, 2, 8, 1, 15, 8, 12, 9, 7, 3, 11, 4, 7, 14, 12])
y = np.array([100, 105, 84, 105, 90, 99, 90, 95, 94, 100, 79, 112, 91, 80, 85])

plt.scatter(x, y)

plt.show()
```

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])
sizes = np.array([10, 25, 50, 100, 250, 500, 30, 45, 5, 150, 300, 100, 40])

plt.scatter(x, y, s=sizes, color='#4CAF50')
plt.show()
```

### Membuat Bar Chart

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width=0.1, color='#4CAF50')
plt.show()
```

### Membuat Histogram

```python
x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

### Membuat Pie Chart

```python
x = np.array([35, 25, 25, 15])

plt.pie(x)
plt.show() 
```

```python
colors = ["cyan", "hotpink", "b", "#4CAF50"]
x = np.array([35, 25, 25, 15])

plt.pie(x, colors = colors)
plt.show() 
```

```python
labels = ["Apples", "Bananas", "Cherries", "Dates"]
explode = [0, 0, 0.2, 0]
x = np.array([35, 25, 25, 15])

plt.pie(x, labels=labels, explode=explode)
plt.legend(loc='upper left')
plt.show() 
```
